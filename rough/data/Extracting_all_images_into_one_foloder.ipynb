```python
from google.colab import drive
drive.mount("/content/drive")

import os
import shutil

BASE_DIR    = "/content/drive/MyDrive/glaze/swatches"
COMBINED    = os.path.join(BASE_DIR, "combined_swatches")
os.makedirs(COMBINED, exist_ok=True)

type_folders = ["type_0", "type_1", "type_2", "type_3", "type_4"]

total = 0
skipped = 0

for type_folder in type_folders:
    type_path = os.path.join(BASE_DIR, type_folder)
    if not os.path.exists(type_path):
        print(f"Skipping {type_folder} — not found")
        continue

    # Each type folder contains subfolders named after the source PDF
    for doc_folder in os.listdir(type_path):
        doc_path = os.path.join(type_path, doc_folder)
        if not os.path.isdir(doc_path):
            continue

        for filename in os.listdir(doc_path):
            if not filename.lower().endswith(".png"):
                continue

            src = os.path.join(doc_path, filename)
            dst = os.path.join(COMBINED, filename)

            # Handle filename collisions by prefixing with type and doc name
            if os.path.exists(dst):
                name, ext = os.path.splitext(filename)
                dst = os.path.join(COMBINED, f"{type_folder}_{doc_folder}_{name}{ext}")
                skipped += 1

            shutil.copy2(src, dst)
            total += 1

print(f"\nDone.")
print(f"Total copied: {total}")
print(f"Renamed due to collisions: {skipped}")
print(f"Combined folder: {COMBINED}")
print(f"Files in combined: {len(os.listdir(COMBINED))}")
```
